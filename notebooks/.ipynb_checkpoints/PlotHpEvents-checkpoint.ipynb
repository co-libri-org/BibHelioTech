```python
import os
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

print(os.getcwd())

import import_ipynb
from CatDfLib import cat2df
```

```python
import sys

filepath = os.path.join(os.getcwd(), '..', 'resources-tests', '105194angeo282332010_bibheliotech_V1.txt')

_df =  cat2df(filepath)




```

```python
start_time = _df.iloc[0,0]
stop_time = _df.iloc[-1,1]

new_idx = pd.date_range(start=start_time, end=stop_time, freq='D')

print(new_idx)

fig, ax = plt.subplots(figsize=(16, 5))
ax.set_title('EUR-CAD rate', fontsize=20)
ax.set_xlabel('Date', fontsize=15)
ax.set_ylabel('Rate', fontsize=15)
ax.set_xlim(_df.index.min(), _df.index.max())
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%b %d'))

date_format = '%d %b %y'
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter(date_format))
ax.set_xticks(new_idx)


# for  _d in _df.values:
#     print(_d)
#     plt.axvspan(_d[0], _d[1], facecolor='yellow', alpha=0.5, hatch='/', edgecolor='red', linewidth=5)
```
