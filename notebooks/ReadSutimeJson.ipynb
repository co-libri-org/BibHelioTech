```python
import os
import json
from pprint import pprint

current_OCR_folder  = "/home/richard/01DEV/bht2/DATA/4F93BC76A241C6B617EFEA50B40C43C9CD72A425/"
current_OCR_folder  = "/home/richard/01DEV/bht2/DATA/web-upload/6EBCE15CF98B3998683104BFA3D0BFF27D19A39E/"

def show_json(file):
    with open(file, "r") as jsonfile:
            mydicts = json.load(jsonfile)
    msg = mydicts.pop()

    # compute types
    all_types = [elmt["type"] for elmt in mydicts]
    uniq_types = sorted(set(all_types))
    count_types = {_t:all_types.count(_t) for _t in uniq_types}

                        
    
    msg = f'{msg}: {len(mydicts)} elmnts'
    msg_sub = len(msg)*"-"
    print(f'{msg_sub:^50}')
    print(f'{msg:^50}')
    print(f'{msg_sub:^50}')
    print("")

    for k, v in count_types.items():
        type_number = f"{v:>3} {k:8}"
        print(f'{type_number:^50}')
    print("\n\n")

    title_str = f'{"type":9}: {"text":27} {"timex-value":>20}      {"value"}'
    print(title_str)
    print(len(title_str)*'-')
    for elmt in mydicts:
        if type(elmt) is not dict or "timex-value" not in elmt:
            continue
        _type = elmt["type"]
        _text = f'"{elmt["text"]}"'
        _timex = elmt["timex-value"]
        _value = elmt["value"]
        print(f'{_type:9}: {_text:27} {_timex:>20} ---> {_value}')


    print("\n\n")
    

for i in range(9):
    show_json(os.path.join(current_OCR_folder,  f"raw{i}_sutime.json"))
```
